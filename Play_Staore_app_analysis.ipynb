## **Data Manipulation**

Since we have app_id, remove id which is as same purpose as app_id.

```python
country_app_data = country_app_data.drop('id',axis = 1)
```

If we notice *'installs'*, it has object data type and the value is with '+' which denotes more than.

> 100000000+  = more than 1 million

So, it has the certaim no of installs without +. Remove + and convert it into numeric data type

```python
country_app_data['installs'] = pd.to_numeric(country_app_data['installs']
                                .astype(str)
                                .str.replace(',' , '', regex = False)
                                .str.replace('+' , '', regex = False),
                                             errors= 'coerce')
```

$f$$r$$e$$e$ is in *object* data type with $T$$r$$u$$e$ & $F$$a$$l$$s$$e$. COnvert it into bool which is appropriate.

```python
country_app_data['free'] =country_app_data['free'].astype(bool)
```

In $c$$o$$u$$n$$t$$r$$y$ column it is showing countries abbrevations. Change abbrevation into full country name to make identify the countries easier.

```python
country_app_data['country'].unique()
```

```python
country_app_data['country'].replace({'us': 'USA',
                                     'in' : 'India',
                                     'br' : 'Brazil',
                                     'id' : 'Indonesia',
                                     'mx' : 'Mexico',
                                     'gb' : 'United Kingdom',
                                     'de' : 'Germany',
                                     'jp' : 'Japan',
                                     'kr' : 'South Korea',
                                     'ph' : 'Philippines'},
                                    inplace = True)
```

```python
country_app_data.info()
```

```python
country_app_data.head()
```

## **Data Cleaning**

```python
country_app_data.isnull().sum()
```

```python
install_per = round(country_app_data['installs'].isnull().mean()* 100.00,2)
min_install_per = round(country_app_data['min_installs'].isnull().mean()* 100.00,2)
price_per = round(country_app_data['price'].isnull().mean()* 100.00,2)

score_per = round(country_app_data['score'].isnull().mean()* 100.00,2)

print(f'Installs null percentage: {install_per}%')
print(f'Min Installs null percentage: {min_install_per}%')
print(f'price null percentage: {price_per}%')

print(f'\nTotal percentage of Install, min_Installs, price :{install_per + min_install_per + price_per}')

print(f'\nScore null percentage: {score_per}%')
```

1. Rows with missing install and price information represented less than 0.3% of the dataset and were removed to preserve data integrity without materially affecting the analysis.

```python
country_app_data.dropna(subset = ['installs','min_installs','price'], inplace = True)
```

2. score, ratings, reviews (37,496 missing values)

These are missing for about 33.7% of the data, so dropping them would remove one-third of your dataset.

> Keep NaN for score, ratings, and reviews because they represent unavailable information rather than zero.

- fillna(0) - If we observe data in those columns they have 0 as data. Filling with 0 will mislead that from unavailable information to 0 rating, score or review.

- fillna('median()') - Score is calculated by average of ratings. Since ratings is also have null values, calculating average of ratings with null values and takinng median is not correct information.

```python
country_app_data.duplicated().sum()
```

```python
country_app_data.drop_duplicates(inplace = True)
```

```python
country_app_data.duplicated().sum()
```

```python
country_app_data.shape
```

```python
country_app_data.to_csv('country_apps.csv', index = False)
```
